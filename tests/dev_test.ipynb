```python
plotter.plot4d_CS(lambda x, y, z: 2*x**2+y+z, 10)
```

```python
import numpy as np
from math import cos, sin
from IPython.display import Image

f = lambda x, y, z: sin(x)*y*cos(z)-x**3
z_range = np.linspace(0,2,10)
frame = plotter.Frame2D(xmin=0, xmax=1, ymin=0, ymax=1)
gif_name = plotter.plot4d(f, z_range, frame=frame, func_name='f', path="gif/")

with open(gif_name,'rb') as f:
    display(Image(data=f.read(), format='png'))
```
