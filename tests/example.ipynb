```python
from plot4d import plotter
```

## Plot a cross-section

```python
plotter.plot4d_CS(lambda x, y, z: x+y+z, 10)
```

## Let create an animation of cross-section plots

```python
gifname = plotter.plot4d(lambda x, y, z: x**2+y-y**3*z, range(0,5), save_images=False)
```

```python
from IPython.display import Image

with open(gifname,'rb') as f:
    display(Image(data=f.read(), format='png'))
```
