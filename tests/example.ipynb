```python
from plot4d import plotter
%matplotlib widget
```

## Create a 4D function $f$

```python
from math import sin, cos
import numpy as np
f = lambda x, y, z: x**2+y-y**3*z
```

## Plot cross-section $f(x, y, z=0)$

```python
plotter.plot4d_CS(f, 0)
```

## Use animation to visualize $f$

```python
gifname = plotter.plot4d(f, np.linspace(0,5,15), func_name="f",fps=5)
```

```python
from IPython.display import Image

with open(gifname,'rb') as file:
    display(Image(data=file.read(), format='png'))
```

## Use `plot4d_data` to check that samples from $f$ show the same structure

```python
np.random.seed(41)

X = np.random.random(size=(500,3))
plotter.plot4d_data(X, np.array([f(x,y,z) for x,y,z in X]))
```
